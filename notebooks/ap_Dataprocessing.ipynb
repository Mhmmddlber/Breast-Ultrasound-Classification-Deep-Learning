```python
import os
import shutil
import random

def resimleri_tasi(kaynak_yolu, hedef_train, hedef_test, hedef_val, oran_train=0.7, oran_test=0.15, oran_val=0.15):

    # Klasörlerin var olduğundan emin olalım, yoksa hata verir
    os.makedirs(hedef_train, exist_ok=True)
    os.makedirs(hedef_test, exist_ok=True)
    os.makedirs(hedef_val, exist_ok=True)

    # Sadece png dosyalarını al
    dosyalar = [f for f in os.listdir(kaynak_yolu) if f.lower().endswith('.png')]
    random.shuffle(dosyalar) # Listeyi karıştır

    toplam_dosya = len(dosyalar)

    # Sayıları hesapla
    sayi_train = int(toplam_dosya * oran_train)
    sayi_test = int(toplam_dosya * oran_test)
    # Kalanların hepsi validation olsun (sayı yuvarlama hatası olmasın diye)
    sayi_val = toplam_dosya - sayi_train - sayi_test

    print(f"Toplam resim: {toplam_dosya}")
    print(f"Train'e gidecek: {sayi_train}")
    print(f"Test'e gidecek: {sayi_test}")
    print(f"Valid'e gidecek: {sayi_val}")

    # --- DİLİMLEME VE TAŞIMA (Kritik Düzeltme Burası) ---

    # 1. Train (%0 - %70 arası)
    train_dosyalari = dosyalar[:sayi_train]
    for dosya in train_dosyalari:
        shutil.move(os.path.join(kaynak_yolu, dosya), os.path.join(hedef_train, dosya))

    # 2. Test (%70 - %85 arası) -> Kaldığımız yerden devam ediyoruz
    test_dosyalari = dosyalar[sayi_train : sayi_train + sayi_test]
    for dosya in test_dosyalari:
        shutil.move(os.path.join(kaynak_yolu, dosya), os.path.join(hedef_test, dosya))

    # 3. Validation (Geriye kalanlar)
    val_dosyalari = dosyalar[sayi_train + sayi_test :]
    for dosya in val_dosyalari:
        shutil.move(os.path.join(kaynak_yolu, dosya), os.path.join(hedef_val, dosya))

    print("Tüm taşıma işlemi başarıyla tamamlandı!")

# --- KULLANIM ---
kaynak = "/content/drive/MyDrive/data/benign"

# Dikkat: Değişken isimleri farklı olmalı
hedef_tr = "/content/drive/MyDrive/data/train/benign"
hedef_ts = "/content/drive/MyDrive/data/test/benign"
hedef_vl = "/content/drive/MyDrive/data/valid/benign"

resimleri_tasi(kaynak, hedef_tr, hedef_ts, hedef_vl)





```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
import os

image_gen = ImageDataGenerator(rotation_range=15,
                               brightness_range=[0.8,1.2],
                               width_shift_range=0.10,
                               height_shift_range=0.10,
                               #rescale=1/255,
                               shear_range=0.1,
                               zoom_range=0.1,
                               horizontal_flip=True,
                               vertical_flip=False,
                               fill_mode='nearest')

for image in os.listdir('/content/drive/MyDrive/data/train/normal'):
  if not image.lower().endswith(('.png', '.jpg', '.jpeg')):
        continue
  try:
    loaded_image = load_img('/content/drive/MyDrive/data/train/normal/' + image, target_size=(224,224))
    x = img_to_array(loaded_image)
    x = x.reshape((1,) + x.shape)

    i = 0
    for batch in image_gen.flow(x, batch_size=1, save_to_dir='/content/drive/MyDrive/data/train/normal', save_prefix='aug', save_format='png'):
        i += 1
        if i >= 25:
            break
  except Exception as e:
        print(f"Bir resimde hata oldu, pas geçiliyor: {image} - Hata: {e}")





```
